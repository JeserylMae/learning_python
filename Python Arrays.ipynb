##### So, we can create an array of integers and float using array module. Let’s get started with the array module and look at all the operations it provides.

```python
import array
# creating array
#INT ARRAY
int_array = array.array('i', [1, 2, 3, 4])
 
#FLOAT ARRAY    
float_array = array.array('f', [1.1, 2.2, 3.3, 4.4])
 
# unicode array support is deprecated and will be deleted in Python 4
unicode_array = array.array('u', ['\u0394', '\u2167', '\u007B'])
```

```python
# printing array
print(int_array)

print(float_array)

print(unicode_array)

print(type(int_array))
```

```python
# We can print array elements using the for loop.
 
int_array = array.array('i', [1, 2, 3, 4])
 
for a in int_array:
    print(a)
```

```python
# We can also access an element using its index. We can use the indices to print the array elements.

for b in range(0, len(int_array)):
    print(f'int_array[{b}] = {int_array[b]}')
```

```python
# Inserting and Appending Elements

# We can use the insert() function to insert an element at the specified index. 
# The elements from the specified index are shifted to right by one position.

int_array = array.array('i', [1, 2, 3, 4])
print(int_array)
```

```python
int_array.insert(0, -1)  # -1,1,2,3,4
print(int_array)
```

```python
int_array.insert(2, -2)  # -1,1,-2,2,3,4
print(int_array)
```

```python
print(int_array)
```

```python
# If you have to add an element at the end of the array, use append() function.

int_array = array.array('i', [1, 2, 3, 4])
```

```python
int_array.append(-3)
```

```python
print(int_array)  # array('i', [1, 2, 3, 4, -3])
```

```python
# Removing Array Elements

# We can use remove() method to remove an array element.

int_array = array.array('i', [1, 2, 3, 4])
print(int_array)
```

```python
# If the element is not present in the array, ValueError is raised.
int_array = array.array('i', [1, 2, 3, 4])
try:
    int_array.remove(20)
except ValueError as ve:
    print(ve)
```

```python
# We can also use pop() function to remove an element at the given index. 
# This function returns the element being removed from the array. 
# If we don’t specify the index, the last element is removed and returned.

int_array = array.array('i', [1, 2, 3, 4])
print(int_array)
```

```python
last_element = int_array.pop()
```

```python
print(last_element)  # 4
```

```python
second_element = int_array.pop(1)
```

```python
# Slicing an Array

# Python array supports slicing and returns a new array with the sub-elements. 
# The original array remains unchanged. The slicing also supports negative indices.

int_array = array.array('i', [0, 1, 2, 3, 4, 5])
print(int_array)
```

```python
print(int_array[3:])  # array('i', [3, 4, 5])
```

```python
print(int_array[:2])  # array('i', [0, 1])
```

```python
print(int_array[1:3])  # array('i', [1, 2])
```

```python
 # negative index slicing
print(int_array[-2:])  # array('i', [4, 5])
```

```python
print(int_array[:-2])  # array('i', [0, 1, 2, 3])
```

```python
# Searching an Element in the Array

int_array = array.array('i', [0, 1, 2, 3, 1, 2])
print(int_array)
```

```python
print(f'1 is found at index {int_array.index(1)}')
try:
    print(int_array.index(20))
except ValueError as ve:
    print(ve)
```

```python
# Updating Value at Specified Index

# We can use the array index with the assignment operator to update the value at an index. 
# If the index is invalid, IndexError is raised.

int_array = array.array('i', [0, 1, 2, 3, 1, 2])
print(int_array)
```

```python
int_array[0] = -1
print(int_array)
```

```python
int_array[1] = -2
print(int_array)
```

```python
try:
    int_array[10] = -10
except IndexError as ie:
    print(ie)
```

```python
# Reversing an Array
# We can use the reverse() function to reverse the array elements.
int_array = array.array('i', [0, 1, 2, 3])
print(int_array)
```

```python
int_array.reverse()
print(int_array)  # array('i', [3, 2, 1, 0])
```

```python
# Count of the Occurrence of an Element

# We can use the count() function to get the number of occurrences of a value in the array.

int_array = array.array('i', [0, 1, 1, 0])
print(int_array)
```

```python
print(int_array.count(1))  # 2
print(int_array.count(0))  # 0
```

```python
# Extending an Array by Appending an Iterable

# We can use the extend() function to append values from the iterable to the end of the array.

array1 = array.array('i', [0, 1])
print('Array 1',array1)


array2 = array.array('i', [2, 3, 4])
print('Array 2:', array2)
```

```python
array1.extend(array2)
```

```python
print(array1)  # array('i', [0, 1, 2, 3, 4])
```

```python
print(array2)  # array('i', [2, 3, 4])
```

```python
array2.extend([1, 2])
print(array2)  # array('i', [2, 3, 4, 1, 2])
```

```python
array1 = array.array('i', [1])
print(array1)
```

```python
array2.extend([1, 2])
print(array2)  # array('i', [2, 3, 4, 1, 2])
```

```python
# Converting Array to List

# We can use the tolist() function to convert an array to a list.

int_array = array.array('i', [0, 1, 2, 3])
print(int_array)
```

```python
print(int_array.tolist())  # [0, 1, 2, 3]
```

Python provides many ways to create 2-dimensional lists/arrays. However one must know the differences between these ways because they can create complications in code that can be very difficult to trace out. 
Let’s start by looking at common ways of creating 1d array of size N initialized with 0s.

```python
# Using 2D arrays/lists the right way'

# Example 1: Creating 1d list Using Naive methods

N = 5
ar = [0]*N
print(ar)
```

```python
# Example 2: creating 1d list using  List Comprehension

N = 5
arr = [0 for i in range(N)]
print(arr)
```

Here we are multiplying the number of rows to the empty list and hence a entire list is created with every element is zero.

```python
#Method 2 Creating a 2-D list

#Example 1: Naive Method'

rows, cols = (5, 5)
arr = [[0]*cols]*rows
print(arr)
```

#### Python has no built-in Arrays because arrays are used for static typing languages

There are two main differences between dynamic typing and static typing that you should be aware of when writing transformation scripts. 

First, dynamically-typed languages perform type checking at runtime, while statically typed languages perform type checking at compile time

In general, statically typed languages will execute faster than dynamically typed ones because their type systems are implemented at compile-time rather than run-time, so all the work involved is performed once only.
